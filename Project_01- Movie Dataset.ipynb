__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).

* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies...

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

__Define__ an appropriate __user-defined function__ to reuse code.

__Movies Top 5 - Highest Revenue__

```python
from IPython.display import HTML
```

```python
movies_best = movies[['poster_path', 'title', 'budget_musd', 'revenue_musd', 'vote_count',  'vote_average', 'popularity']].copy()
movies_best
```

```python
movies_best['profit_usd'] = movies_best.revenue_musd.sub(movies_best.budget_musd)
movies_best['return'] = movies_best.revenue_musd.div(movies_best.budget_musd)
movies_best.head()
```

```python
# Change column names for data presentation purposes
movies_best.columns = ['', 'Title', 'Budget', 'Revenue', 'Votes', 'Average Rating', 'Popularity', 'Profit', 'Return']
movies_best.columns
```

```python
movies_best.set_index('Title', inplace = True)
movies_best.head()
```

```python
movies_best.iloc[1,0]
```

```python
subset = movies_best.iloc[:5, :2]
subset
```

```python
# Converting to html code
subset.to_html(escape = False)
```

```python
# converting from raw data to html front end display
HTML(subset.to_html(escape = False))
```

```python
# Function for producing best or worst movies based on the parameter 'by'

def best_worst(num, by, ascending = False, min_budget = 0, min_votes = 0):
    """Filters for best movies based on the parameter 'by' if ascending = False and
       Filters for worst movies based on the parameter 'by' if ascending = True"""
    df = movies_best.loc[(movies_best.Budget >= min_budget) & (movies_best.Votes >= min_votes),
                         ['', by]].sort_values(by = by, ascending = ascending).head(num).copy()
    return HTML(df.to_html(escape = False))
```

__Movies Top 5 - Highest Budget__

```python
best_worst(5, 'Budget')
```

__Movies Top 5 - Highest Profit__

```python
best_worst(5, 'Profit')
```

__Movies Top 5 - Lowest Profit__

```python
best_worst(5, 'Profit', ascending = True)
```

__Movies Top 5 - Highest ROI__

```python
best_worst(5, 'Return', min_budget = 5)
```

__Movies Top 5 - Lowest ROI__

```python
best_worst(5, 'Return', min_budget = 50, ascending = True)
```

__Movies Top 5 - Most Votes__

```python
best_worst(5, 'Votes', min_budget = 5)
```

__Movies Top 5 - Highest Rating__

```python
best_worst(5, 'Average Rating', min_votes = 100)
```

__Movies Top 5 - Lowest Rating__

```python
best_worst(5, 'Average Rating', min_votes = 100, ascending = True)
```

__Movies Top 5 - Most Popular__

```python
best_worst(5, 'Popularity')
```
