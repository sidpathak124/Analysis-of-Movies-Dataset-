### Orientations of JSONs: Records, columns, split. Above we saw record.
record: List of dicts. Each dict in the list represent an observation (movie here)

columns: Dict of dicts. Each dict in the dict represent a single feature for all observations

split: Dict with 3 lists. Each list as value for index, columns and data

### There are pros and cons of each orientation

split: takes less memory but is error prone;
records: takes more memory but is better compatible with pandas

```python
movie_id = list(df3.id)
movie_id
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}'
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)
    r = requests.get(url)
    if r.status_code != 200:
        continue
    else:
        data = r.json()
        json_list.append(data)
df = pd.DataFrame(json_list)
df
        
```

```python
df = df.loc[:, ['title', 'id', 'revenue', 'genres', 'belongs_to_collection', 'runtime']].sort_values(by = 'revenue', ascending = False)
df
# Most successful movies by revenue (not adjusted for inflation)
```
